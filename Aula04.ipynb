```python
#Criando uma lista em Python
list = [0, 'eu amo polAtm', 2.5]
print (list)
print ('variável: '+list[1]) #pedi para mostrar o segundo elemento, que é a string. Por isso, permite somar com o texto que é outra string
print(list[0]+1) #permite somar tipos iguais, neste caso um numero inteiro com outro inteiro
```

```python
#Criando uma matriz numpy
import numpy as np #importa a funçao, mas para isso ela precisa estar instalada (pra instalar, precisa ir no terminar e colocar: pip install numpy - por exemplo)
#pra aparecer as funcoes no np, coloca np. e da tab
a = np.zeros((5,5)) #cria uma matriz de zeros utilizando o comando próprio np.zeros
print(a)

print('Indexação numpy')
print(a[0,0]) #primeiro é a linha, depois a coluna; sempre considerando que a contagem inicia em 0 tanto pra linha quanto pra coluna
```

```python
#Criando números aleatórios (entre 0 1 e) e mostrando a dimensão da matriz, que inicia em 0 e vai até 4.
b = np.random.rand(5,5) #cria uma matriz de numeros aleatorios
print (b)
print(b.shape) #é o shape/dimensão da matriz
```

```python
#Para fazer a matriz aleatória ficar entre dois valores, faz: Vmin+(Vmax - Vmin) * rand - chamado Metodo Monte Carlos
#Criando matriz de valores aeatórios entre uma faixa estabelecida
vmin = 1
vmax = 100

c = vmin + (vmax-vmin)*b
print(c)
```

```python
#Plotar matriz
import matplotlib.pyplot as plt

plt.plot(c.transpose()) #plota cada linha e suas colunas
```

```python
#Plotando apenas a primeira linha
plt.plot(c[0,:])#linha 0, todas as colunas
```

```python
x = np.array(range(0,np.size(c[0,:]))) #np.array converte o range (0,5) em um dado tipo numpy
print(x)
plt.plot(c[0,:])
```

```python
x = np.array(range(0,np.size(c[0,:]))) #np.array converte o range (0,5) em um dado tipo numpy
print(x)
plt.plot(x, c[0,:])
plt.title('Gráfico com valores inteiros no eixo x')
plt.xticks(x)
```

```python
#Criando um gráfico de barras e exportando

fig, ax = plt.subplots(3,1)
ax[0].bar(x,c[:,0]) #ax é um objeto plt
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

#para exportar
import os
rootFolder = os.getcwd()
figFolder = rootFolder.replace('\\','/')+'/figures/'
print(figFolder)
print(rootFolder)
fig.savefig(figFolder+'Figure01.png') 
```
